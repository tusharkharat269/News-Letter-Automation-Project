```python
import os
from dotenv import load_dotenv
load_dotenv()

BASE_URL = "https://openrouter.ai/api/v1"
API_KEY = os.getenv("OPENROUTER_API_KEY")
```

```python
from langchain_openai import ChatOpenAI

# DeepSeek model hosted on OpenRouter
llm = ChatOpenAI(
    model="deepseek/deepseek-chat",  # or "deepseek/deepseek-r1"
    temperature=0.7,
    api_key= API_KEY,
    base_url=BASE_URL,
)

# response = llm.invoke("hi")
# print(response.content)
```

```python
load_dotenv()

tavily_key = os.getenv("TAVILY_API_KEY")
# print(tavily_key)
```

```python
from langchain_tavily import TavilySearch

search = TavilySearch(max_results=15,time_range="week")
```

```python
query="technology news"
results = search.run(query)
```

```python
import json

print(json.dumps(results,indent=2))
```

```python
from langchain.prompts import PromptTemplate
from langchain.chains.llm import LLMChain

prompt = PromptTemplate(
    input_variables=["query", "results"],
    template="""
You are a Technology news research assistant.

Task:
- Analyze the following search results: {results}
- Select the TOP 5 most important technology news.
- For each, return JSON with:
  - headline
  - short_summary (2â€“3 sentences)
  - source_url

Query: {query}
---
Return only valid JSON list of 5 items in JSON format only.
"""
)
```

```python
# print(results["results"])

# print(type(results["results"]))

res = str(results["results"])
print(type(res))
```

```python
news_chain = LLMChain(llm=llm, prompt=prompt)


raw_news = news_chain.invoke({"query": query, "results": results})
```
